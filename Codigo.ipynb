```python
# 1. Definición de las Clases
class Empresa:
    """Clase para representar a la empresa cliente"""
    def __init__(self, nombre, sector):
        self.nombre = nombre
        self.sector = sector
    
    def __str__(self):
        return f"Empresa: {self.nombre} (Sector: {self.sector})"

class Campana:
    """Clase para representar la campaña de marketing"""
    def __init__(self, objeto_empresa, presupuesto):
        self.empresa = objeto_empresa  # Relación entre clases
        self.presupuesto = presupuesto
        self.estado = "Protegido (Escrow)"
    
    def mostrar_detalle(self):
        print(f"--- Detalle de la Campaña ---")
        print(f"Cliente: {self.empresa.nombre}")
        print(f"Presupuesto: ${self.presupuesto}")
        print(f"Estado de seguridad: {self.estado}")

# 2. Instanciación 
mi_empresa = Empresa("Sebastian Corp", "Marketing Digital")
mi_proyecto = Campana(mi_empresa, 500)

# 3. Prueba de funcionamiento 
print(mi_empresa)
mi_proyecto.mostrar_detalle()
```
